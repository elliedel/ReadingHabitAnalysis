Based on my observation, the NaN values on Bookshelves reprents the books I already read.
Best way to approach for me imo is to delete all the "reading" and "to be read" books since I'm only interested on books I already read. An indicator that a book is already read is if the book does not have a date read value AND the value in Bookshelves in "NaN". The reason is that some books that I read doesn't have a date read because I already read those books prior to having a Goodreads account.

```python
my_library[(my_library["Bookshelves"].notna()) & (my_library["Date Read"].isna())]
```

based on checking alone, I conclude that these books are the ones I have not read before. Time to drop them.

```python
# gets the indices of the rows that does not satisfy the conditions
my_library.drop(my_library[(my_library["Bookshelves"].notna()) & (my_library["Date Read"].isna())].index, inplace=True)

my_library.info()
```

```python
print("New Datset Shape: ", my_library.shape)
```

```python
my_library.head()
```

```python
# drop unneccesary columns
my_library.drop(columns=[
   "Author l-f", 
   "ISBN13", "Binding", 
   "Year Published", 
   "Bookshelves with positions",
   "Exclusive Shelf",
   "My Review",
   "Spoiler",
   "Private Notes",
   "Owned Copies",
   "Book Id",
   "Bookshelves",
   "Publisher",
   "Average Rating",
    ], inplace=True)
```

```python
my_library.info()
```

There are still a lot of rows with null values. Additional Authors can be disregarded completely, as well as Date Read column due to some books being read prior to having a Goodreads Account.

```python
print("Book without the Number of Pages")
my_library[my_library["Number of Pages"].isna()]
```

```python
print("Book without Original Publication Year")
my_library[my_library["Original Publication Year"].isna()]
```

Based on the filter, these two books are not books that we do not need for the analysis, due the nature of the queried books (Web comic and web novel). 

```python
my_library.drop(my_library[my_library["Original Publication Year"].isna()].index, inplace=True)
```
